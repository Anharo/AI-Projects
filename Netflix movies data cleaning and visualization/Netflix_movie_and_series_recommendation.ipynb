```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix_data = pd.read_csv('netflix_titles.csv')
netflix_data.head()
```

```python
netflix_data_empty = netflix_data.isna().sum()
netflix_data_empty
```

```python
netflix_data[['director', 'cast', 'country', 'duration']] = netflix_data[['director', 'cast', 'country', 'duration']].fillna('')
```

```python
netflix_data.dropna(subset=['date_added','rating'], inplace=True)
```

```python
netflix_data.head()
```

```python
netflix_data_empty = netflix_data.isna().sum()
netflix_data_empty
```

```python
a = netflix_data.shape
a
```

```python
netflix_data.drop_duplicates(inplace=True)
```

```python
a = netflix_data.shape
a
```

```python
netflix_data.info()
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,8))
type_value = netflix_data['type'].value_counts()
sns.countplot(data = netflix_data, x = 'type', palette = 'viridis', ax = axes[0])
axes[0].set_title('Types of Content')
axes[1].pie(type_value, labels = type_value.index)
axes[1].set_title('Types of Content')
plt.show()
```

```python
director_film = netflix_data[netflix_data["type"] == "TV Show"]["director"].value_counts()[1:11]
plt.figure(figsize = (12,8))
sns.barplot(x = director_film.index, y = director_film.values)
plt.xticks(rotation = 45, fontsize = 14)
plt.title("Top 10 Directors with the Most Films",fontsize = 18)
plt.show()
```

```python
netflix_data['director'].value_counts()
```

```python
top_country = netflix_data['country'].replace('', np.nan).dropna().value_counts().head(10)
top_country
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x = top_country.index, y = top_country.values)
plt.title('Top 10 Countries with the Most Content', fontsize = 18)
plt.show()
```

```python
netflix_data["date_added"] = pd.to_datetime(netflix_data["date_added"], errors='coerce')
netflix_data["year_added"] = netflix_data["date_added"].dt.year
netflix_data["month_added"] = netflix_data["date_added"].dt.month
```

```python
release_counts = netflix_data.groupby(["release_year", "type"]).size().reset_index(name="count")

sns.lineplot(data=release_counts, x="release_year", y="count", hue="type")
plt.title("Number of TV Series and Movies Released Each Year", fontsize = 16)
plt.show()
```
